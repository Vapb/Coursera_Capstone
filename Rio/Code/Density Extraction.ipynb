```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

# WebScraping data from Neighbourhoods from RIO (Population Density)

```python
url_bairros = 'https://pt.wikipedia.org/wiki/Lista_de_bairros_da_cidade_do_Rio_de_Janeiro'
```

```python
source = requests.get(url_bairros).text
soup = BeautifulSoup(source,'lxml')
```

```python
# Find table from wiki
table = soup.find('table', class_='wikitable')
trs = table.find_all('tr')
```

```python
# Find All Links
hrefs = []
for tr in trs:
    for td in tr.find_all('td'):
        for a in td.find_all('a'):
            href = a.get('href')
            url = 'https://pt.wikipedia.org' + href
            if href[0] != '#' and not('(estado)' in href) and not('.JPG' in href) and not('.jpg' in href) :
                hrefs.append(url)
```

```python
hrefs[0:5]
```

### Remove Unwanted Links 

```python
hrefs.remove('https://pt.wikipedia.org/wiki/Zona_Sul_(Rio_de_Janeiro)')
hrefs.remove('https://pt.wikipedia.org/wiki/Zona_Sul_(Rio_de_Janeiro)')
hrefs.remove('https://pt.wikipedia.org/wiki/Zona_Oeste_(Rio_de_Janeiro)')
hrefs.remove('https://pt.wikipedia.org/wiki/Barra_e_Jacarepagu%C3%A1')
hrefs.remove('https://pt.wikipedia.org/wiki/Grande_Bangu')
hrefs.remove('https://pt.wikipedia.org/wiki/Ilha_do_Governador')
hrefs.remove('https://pt.wikipedia.org/wiki/Subprefeitura_da_Zona_Oeste_(Rio_de_Janeiro)')
hrefs.remove('https://pt.wikipedia.org/wiki/Zona_Norte_(Rio_de_Janeiro)')
hrefs.remove('https://pt.wikipedia.org/wiki/Grande_Tijuca')
hrefs.remove('https://pt.wikipedia.org/wiki/Grande_M%C3%A9ier')
hrefs.remove('https://pt.wikipedia.org/wiki/Zona_da_Leopoldina')
hrefs.remove('https://pt.wikipedia.org/wiki/Subprefeitura_da_Zona_Norte_da_cidade_do_Rio_de_Janeiro')
```

## Extract Population and Area from Neighbourhoods Links

```python
result = []

for x in hrefs:
    print(x)
    source = requests.get(x).text
    soup = BeautifulSoup(source,'lxml')
    table = soup.find('table', class_="infobox_v2")
    if table == None:
        table = soup.find('table', class_="infobox")
        st = "vertical-align:top; text-align:left;"
    else:
        st = "vertical-align: top; text-align: left;"

        
    hab = '0.0'
    area = '0,0'
    if table != None: 
        table.find_all('span', class_='fn')
        trs = table.find_all('tr')
        flag = True
        for tr in trs:
            for td in tr.find_all('td', style=st):
                information = td.text
                if flag and 'hab' in information:
                    hab = information[0:7]
                    flag = False
                elif flag and '2010' in information:
                    hab = information[0:7]
                    flag = False
                elif 'km' in information:
                    area = information[0:12]
                elif '2003' in  information:
                    area = information[0:10]
                
    row = [hab, area]
    print(row)
    result.append([hab,area]) 
```

## Results into DataFrame

```python
resultdf = pd.DataFrame(result)
```

```python
resultdf.rename(columns={0 : 'Hab', 1 : 'Area'}, inplace=True)
```

```python
resultdf.head()
```

# CLEAN DATA

```python
resultdf['Hab'] = resultdf['Hab'].str.strip(' (em')
resultdf['Hab'] = resultdf['Hab'].str.replace('\\xa0',' ')
resultdf['Hab'] = resultdf['Hab'].str.replace(',','')
resultdf['Hab'] = resultdf['Hab'].str.replace(' ','')
resultdf['Hab'] = resultdf['Hab'].str.replace('.','')
```

```python
resultdf['Area'] = resultdf['Area'].str.replace('.','')
resultdf['Area'] = resultdf['Area'].str.replace(',','.')
resultdf['Area'] = resultdf['Area'].str.replace('ha','')
resultdf['Area'] = resultdf['Area'].str.replace('./','')
resultdf['Area'] = resultdf['Area'].str.replace('\\xa0','')
resultdf['Area'] = resultdf['Area'].str.replace('h','')
resultdf['Area'] = resultdf['Area'].str.replace('(','')
resultdf['Area'] = resultdf['Area'].str.replace('em','')
resultdf['Area'] = resultdf['Area'].str.replace('kmÂ²','')
resultdf['Area'] = resultdf['Area'].str.replace(' ','')
resultdf['Area'] = resultdf['Area'].str.replace('[','X')
resultdf['Area'] = resultdf['Area'].str.replace(']','X')
resultdf['Area'] = resultdf['Area'].str.replace('X1XX2X','')
resultdf['Area'] = resultdf['Area'].str.replace('ec','')
```

## STRING TO FLOAT

```python
resultdf['Area'] =  resultdf['Area'].apply(lambda x: float(x))
resultdf['Hab'] =  resultdf['Hab'].apply(lambda x: int(x))
```

```python
import folium
from branca.colormap import linear
```

**COLOR MAP Hab**

```python
mini = merged['Hab'].min()
maxi = merged['Hab'].max()
z = (mini+maxi)/2
za = (z + mini)/2
zb = (z + maxi)/2
colormap = linear.Greens_09.scale(mini,maxi)
colormap = colormap.to_step(index=[500,37500,75000,150000,300000])
colormap
```

# Total Population Map

```python
# create map of New York using latitude and longitude values
map_rj = folium.Map(location=[latitude, longitude], zoom_start=11)

# add markers to map
for lat, lng, bairro, hab in zip(merged['LAT'], merged['LON'], merged['BAIRRO'], merged['Hab']):
    label = '{}, {}'.format(bairro, hab)
    label = folium.Popup(label, parse_html=True)
        
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='black',
        fill=True,
        fill_color= colormap(hab),
        fill_opacity=0.7,
        parse_html=False).add_to(map_rj)  


colormap.caption = 'HAB NEIGHBOURHOOD RJ'
colormap.add_to(map_rj)

    
map_rj
```
